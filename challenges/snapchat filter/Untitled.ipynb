```python
import cv2 
import numpy as np
import os
import matplotlib.pyplot as plt
import pandas as pd
```

```python
eye_cascade = cv2.CascadeClassifier("Train/third-party/frontalEyes35x16.xml")
```

```python
glasses = cv2.imread("Train/glasses.png",cv2.IMREAD_UNCHANGED)
glasses = cv2.cvtColor(glasses,cv2.COLOR_BGR2RGBA)
plt.imshow(glasses)
```

```python
kingslayer = cv2.imread("Train/Jamie_Before.jpg",cv2.IMREAD_UNCHANGED)
kingslayer = cv2.cvtColor(kingslayer,cv2.COLOR_BGR2RGBA)
print(kingslayer.shape)
plt.imshow(kingslayer)
```

```python
eyes = eye_cascade.detectMultiScale(kingslayer,8,5)
print(len(eyes))
x = eyes[0][0]
y = eyes[0][1]
w = eyes[0][2]
h = eyes[0][3]
new_glasses = cv2.resize(glasses ,dsize = (w,h))
plt.imshow(new_glasses)
```

```python
alpha = new_glasses[:,:,3]/255
inverted_alpha = 1 - alpha
plt.imshow(inverted_alpha)
```

```python
(X,Y) = alpha.shape

for i in range(X):
    for j in range(Y):
        if(alpha[i,j]):
            kingslayer[y+i,x+j,:3] = new_glasses[i,j,:3]
```

```python
chain = cv2.imread("Train/Gold-Chain.png",cv2.IMREAD_UNCHANGED)
chain = cv2.cvtColor(chain,cv2.COLOR_BGR2RGBA)
plt.imshow(chain)
```

```python
alpha2 = new_chain[:,:,3]/255
(X2,Y2) = alpha2.shape
print(kingslayer.shape)
l=kingslayer.shape[0]
b = kingslayer.shape[1]
```

```python
for i in range(X2):
    for j in range(Y2):
        if(alpha2[j,i]):
            if(y2+h2+j<l and x+i<b):
                kingslayer[y2+h2+j,x2+i,:3] = new_chain[j,i,:3]
```
