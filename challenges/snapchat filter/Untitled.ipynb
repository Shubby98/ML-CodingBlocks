```python
import cv2 
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
cam = cv2.VideoCapture(0)
```

```python
nose_cascade = cv2.CascadeClassifier("Train/third-party/Nose18x15.xml")
```

```python
while True:
    ret, frame = cam.read()
    if ret==False:
        print("kuch to galat ho rha hai")
        continue
    
    key_pressed = cv2.waitKey(1)&0xFF #Bitmasking to get last 8 bits
    if key_pressed==ord('q'): #ord-->ASCII Value(8 bit)
        break
    
    noses = nose_cascade.detectMultiScale(frame,1.3,5)
    for (x,y,w,h) in noses:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),3)
        break
    
    cv2.imshow('Nose Detector',frame)
    
cam.release()
cv2.destroyAllWindows()
    
```
