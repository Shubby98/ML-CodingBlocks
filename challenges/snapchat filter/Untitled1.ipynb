```python
import cv2 
import numpy as np
import os
import matplotlib.pyplot as plt
import pandas as pd
```

```python
eye_cascade = cv2.CascadeClassifier("Train/third-party/frontalEyes35x16.xml")
glasses = cv2.imread("Train/glasses.png",cv2.IMREAD_UNCHANGED)
glasses = cv2.cvtColor(glasses,cv2.COLOR_BGR2RGBA)
kingslayer = cv2.imread("Test/Before.png",cv2.IMREAD_UNCHANGED)
kingslayer = cv2.cvtColor(kingslayer,cv2.COLOR_BGR2RGBA)
print(kingslayer.shape)
eyes = eye_cascade.detectMultiScale(kingslayer,1.3,5)
x = eyes[0][0]
y = eyes[0][1]
w = eyes[0][2]
h = eyes[0][3]
new_glasses = cv2.resize(glasses ,dsize = (w,h))
alpha = new_glasses[:,:,3]/255
inverted_alpha = 1 - alpha
print(eyes)


nose_cascade = cv2.CascadeClassifier("Train/third-party/Nose18x15.xml")
mustache = cv2.imread("Train/mustache.png",cv2.IMREAD_UNCHANGED)
mustache = cv2.cvtColor(mustache,cv2.COLOR_BGR2RGBA)
#plt.imshow(mustache)
noses = nose_cascade.detectMultiScale(kingslayer,4,5)
x1 = noses[0][0]
y1 = noses[0][1]
w1 = noses[0][2]
h1 = noses[0][3]
plt.figure(figsize=(7,7))
#plt.imshow(kingslayer[y1:y1+h1,x1:x1+w1])
new_mustache = cv2.resize(mustache ,dsize = ((23*w1//16),(15*h1//16)))
alpha1 = new_mustache[:,:,3]/255
inverted_alpha1 = 1 - alpha1
#plt.imshow(inverted_alpha1)
```
