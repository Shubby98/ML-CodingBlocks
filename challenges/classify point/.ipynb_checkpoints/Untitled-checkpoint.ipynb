```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = pd.read_csv("Training Data/Logistic_X_Train.csv")
Y = pd.read_csv("Training Data/Logistic_Y_Train.csv")

```

```python
X_train = np.array(X.values[:1500,:])
Y_train = np.array(Y.values[:1500,:])

print(X_train.shape , Y_train.shape)

X_test = np.array(X.values[1500:,:])
Y_test = np.array(Y.values[1500:,:])

print(X_test.shape , Y_test.shape)
```

```python
plt.scatter(X_train[:,0],X_train[:,1],c = Y_train[:,0])
```

```python
plt.scatter(X_test[:,0],X_test[:,1],c = Y_test[:,0])
```

# Building a Neural Network

```python
from keras import models
from keras.layers import Dense
```

```python
model = models.Sequential()
model.add(Dense(16 ,activation = 'relu' ,input_shape = (2,)))
model.add(Dense(16 ,activation = 'relu'))
model.add(Dense(1, activation = 'sigmoid'))
```

```python
model.compile(optimizer='rmsprop',loss = 'binary_crossentropy', metrics =['accuracy'] )
```

```python
model.summary()
```

```python
hist = model.fit(X.values,Y.values,epochs = 500 , batch_size = 512 ,validation_data=(X_train,Y_train) )
```

```python
h =  hist.history
#plt.plot(h['val_loss'] ,label = 'validation loss')
plt.plot(h['loss'] , label = 'training loss' )

plt.legend()
plt.show()
```

```python
h =  hist.history
#plt.plot(h['val_acc'] ,label = 'validation accuracy')
plt.plot(h['acc'] , label = 'training accuracy' )

plt.legend()
plt.show()
```
